```python
#While Loops
##Iterate over the sequence only if condition is met

number = 0

while number < 5:
    print(number)
    number = number + 1
```

```python
number = 0

while number < 5:
    print(number)
    if number == 3:
        break
    number = number + 1
```

```python
number = 0

while number < 5:
    print(number)
    if number == 6:
        break
    number = number + 1
else:
    print('No longer less than 5')
```

```python
##Infinite loop
number = 0

while number < 5:
    print(number)
    if number == 3:
        continue
    number = number + 1
else:
    print('No longer less than 5')
```

```python
number = 0

while number < 5:
    number = number + 1
    if number == 3:
        continue
    print(number)
    
else:
    print('No longer less than 5')
```

```python
#*List, Set & Dictionary Comprehension*

#If you have a operation such as the below 
#numbers = []
#for i in range(1,6):
#    numbers.append(2**i)

a = [2,4,5,8,11]
##lets say you want [4,8,10,16,22]
```

```python
##one way to perform 
b = []
b.append(10)
b.append(20)
print(b)
```

```python
##second way
c = []
for x in a:
    c.append(x*2)
print(c)
```

```python
##list comprehension method
d = [x * 2 for x in a]
print(d)
```

```python
## another example [1,4,9,16,25,36]
```

```python
el = []

for x in range(1,7):
    el.append(x**2)
print(el)
```

```python
##list comprehension
e2 = [x**2 for x in range(1,7)]
print(e2)
```

```python
## suppose the opposite [36, 25, 16, 9, 4, 1]
for x in range(6, 0, -1):
    print(x)
```

```python
f1 = []
for x in range(6, 0, -1):
    f1.append(x**2)
print(f1)
```

```python
f2 = [x**2 for x in range(6,0,-1)]
print(f2)
```
